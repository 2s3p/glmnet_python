## Linear Regression: Gaussian family

We will demonstrate the use of glmnet for the ordinary Gaussian linear regression case. 

**gaussian** is the default family option in the function **glmnet**. Suppose we have observations $x_i \in \mathbb{R}^p$ and the responses $y_i \in \mathbb{R}, i = 1,...,N$. The objective function for the Gaussian family is:

$$ \textrm{min} \frac{1}{2N} \Sigma$$

```python
# Import relevant modules and setup for calling glmnet

%matplotlib inline

import sys
sys.path.append('../test')
sys.path.append('../lib')

import scipy
import glmnet 
import importlib
import pprint
import glmnetPlot
import glmnetPredict
import cvglmnet
import cvglmnetPlot
import glmnetPrint

importlib.reload(glmnet)
importlib.reload(glmnetPlot)    
importlib.reload(glmnetPredict)    
importlib.reload(cvglmnet)    
importlib.reload(cvglmnetPlot)
```

Setup parameters for where the data is, and the model type to be used. 

```python
# parameters
baseDataDir= '../data/'
testType = 'gaussian'
```

Load the data, and display a slice of the data to illustrate the the dataset.

```python
#  elnet caller 
x = scipy.loadtxt(baseDataDir + 'QuickStartExampleX.dat', dtype = scipy.float64)
y = scipy.loadtxt(baseDataDir + 'QuickStartExampleY.dat', dtype = scipy.float64)
print('==========================================')
print('x data:')
print(x.shape)
print(x[0:5, 0:5])
print('==========================================')
print('y data:')
print(y.shape)
print(y[0:5, ])
print('==========================================')
```

Call glmnet directly (without using the cross-validation calls). Plot the results from the glmnet fits. All options are set to default values for this example:

```python
# glmnet, glmnetPlot, glmnetPredict
# create weights
r1 = scipy.ones((50, 1), dtype = scipy.float64)
weights = scipy.row_stack((r1, 2*r1))
# fit
fit = glmnet.glmnet(x = x.copy(), y = y.copy(), family = testType, \
                    weights = weights, \
                    alpha = 0.2, nlambda = 20
                    )

glmnetPrint.glmnetPrint(fit)
```

```python
glmnetPlot.glmnetPlot(fit, xvar = 'lambda', label = True)
```

```python
glmnetPlot.glmnetPlot(fit, xvar = 'dev', label = True)
```

```python
glmnetPredict.glmnetPredict(fit, x[0:5,:], ptype = 'response', \
                                s = scipy.float64([0.05]))
```

```python
# cvglmnet, cvglmnetPlot
fit = cvglmnet.cvglmnet(x = x.copy(), y = y.copy(), family = testType)
cvglmnetPlot.cvglmnetPlot(fit)
```
